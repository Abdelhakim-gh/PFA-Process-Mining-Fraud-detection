# Event Logs To CSV Format

```python
import pm4py
import pandas as pd
from pathlib import Path

# Define the file paths using pathlib
input_file_path = Path('./dataset/Hospital Billing - Event Log.xes.gz')
output_file_path = Path('./dataset/Hospital Billing - Event Log.csv')

# Read the XES log file
log = pm4py.read_xes(str(input_file_path))  # Input Filename

# Convert the log to a Pandas DataFrame
df = pm4py.convert_to_dataframe(log)

# Display the DataFrame
print(df.head())

# Optionally, save the DataFrame to a CSV file
df.to_csv(str(output_file_path), index=False)
```
